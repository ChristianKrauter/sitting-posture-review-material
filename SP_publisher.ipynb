```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

##
## color from https://www.nature.com/articles/nmeth.1618
##
palette = sns.color_palette(
    palette=[
        "#56B4E9",
        "#009E73",
        "#E69F00",
        "#F0E442",
        "#0072B2",
        "#D55E00",
        "#CC79A7",
    ]
)

sns.set_palette(palette)
sns.palplot(pal=palette)
```

```python
df = pd.read_csv("SP_publisher.csv", sep=",")
df = df[["publisher plot", "count plot"]].dropna().drop(labels=6)

fig, axs = plt.subplots(figsize=(6, 4))

plt.barh(width=df["count plot"], y=df["publisher plot"])

plt.xlabel("number of publications", fontsize=18)
axs.tick_params(axis="both", which="major", labelsize=1)

plt.tight_layout()
plt.savefig("SP_publisher.pdf")
plt.savefig("SP_publisher.png", facecolor="white")
plt.show()
```
