```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

##
## color from https://www.nature.com/articles/nmeth.1618
##
palette = sns.color_palette(
    palette=[
        "#56B4E9",
        "#009E73",
        "#E69F00",
        "#F0E442",
        "#0072B2",
        "#D55E00",
        "#CC79A7",
    ]
)

sns.set_palette(palette)
sns.palplot(pal=palette)
```

```python
df = pd.read_csv("SP_areas_overview.csv", sep=",")
df = df[["year", "area"]].groupby(["area"])
df = df.count().reset_index().sort_values(by="year", ascending=False)

fig, axs = plt.subplots(figsize=(7.117, 4))

plt.barh(width=df["year"], y=df["area"])
plt.xlabel("number of publications", fontsize=18)

axs.tick_params(axis="both", which="major", labelsize=18)

fig.tight_layout()
plt.savefig("SP_areas_overview.pdf")
plt.savefig("SP_areas_overview.png", facecolor="white")
plt.show()
```
